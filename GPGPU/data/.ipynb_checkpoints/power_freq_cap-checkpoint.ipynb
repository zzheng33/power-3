```python
import subprocess
import time
import matplotlib.pyplot as plt
import csv
import os

# Available frequencies
frequencies = [
    1410, 1395, 1380, 1365, 1350, 1335, 1320, 1305, 1290, 1275, 1260,
    1245, 1230, 1215, 1200, 1185, 1170, 1155, 1140, 1125, 1110, 1095,
    1080, 1065, 1050, 1035, 1020, 1005, 990, 975, 960, 945, 930, 915, 900
]

# Uniformly select 15 frequencies
selected_freqs = [frequencies[i] for i in 
                  sorted([int(i * (len(frequencies) - 1) / 14) for i in range(15)])]

# Create output directory
output_dir = "gpu_power_logs"
os.makedirs(output_dir, exist_ok=True)
cumulative_file = os.path.join(output_dir, "gpu_power_all.csv")

# Reset cumulative file if it exists
if os.path.exists(cumulative_file):
    os.remove(cumulative_file)

gpu_power_data = {}

for freq in selected_freqs:
    print(f"\nRunning for GPU max frequency cap = {freq} MHz")

    # Set GPU max graphics frequency (min = 210 MHz, max = current freq)
    subprocess.run(f"sudo nvidia-smi -ac 1215,{freq}", shell=True, check=True)
    subprocess.run("sudo nvidia-smi -pm 1", shell=True, check=True)

    # Start raytracing workload
    proc = subprocess.Popen("/home/cc/power/GPGPU/script/run_benchmark/altis_script/level2/raytracing.sh", shell=True)

    power_list = []
    try:
        while proc.poll() is None:
            # Query GPU power
            result = subprocess.check_output(
                "nvidia-smi --query-gpu=power.draw --format=csv,noheader,nounits",
                shell=True
            )
            power = float(result.decode("utf-8").strip().split("\n")[0])
            power_list.append(power)
            time.sleep(1)
    except KeyboardInterrupt:
        proc.terminate()
        break

    gpu_power_data[freq] = power_list

    # Save to individual CSV
    with open(os.path.join(output_dir, f"gpu_power_{freq}.csv"), "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Time (s)", "Power (W)"])
        for i, power in enumerate(power_list):
            writer.writerow([i, power])

    # Append to cumulative CSV
    with open(cumulative_file, "a", newline="") as f:
        writer = csv.writer(f)
        if freq == selected_freqs[0]:  # Write header only once
            writer.writerow(["Frequency (MHz)", "Time (s)", "Power (W)"])
        for i, power in enumerate(power_list):
            writer.writerow([freq, i, power])

# Reset GPU clock caps
subprocess.run("sudo nvidia-smi -rgc", shell=True)
```

```python
log_dir = "./gpu_power_logs/raytracing/"
# Loop through all files that match the pattern
for filename in os.listdir(log_dir):
    if filename.startswith("gpu_power_") and filename.endswith(".csv") and filename != "gpu_power_all.csv":
        freq = int(filename.split("_")[2].split(".")[0])
        filepath = os.path.join(log_dir, filename)

        with open(filepath, "r") as f:
            reader = csv.reader(f)
            next(reader)  # Skip header
            powers = [float(row[1]) for row in reader]

        gpu_power_data[freq] = powers

# Plotting
plt.figure(figsize=(10, 6))
for freq in sorted(gpu_power_data.keys()):
    power_values = gpu_power_data[freq]
    x = [freq] * len(power_values)
    plt.scatter(x, power_values, label=f"{freq} MHz", s=10)

plt.xlabel("GPU Graphics Clock Frequency (MHz)")
plt.ylabel("GPU Power Consumption (W)")
plt.title("GPU Power vs Frequency")
plt.grid(True)
plt.tight_layout()
plt.savefig("./fig/gpu_power_vs_frequency_from_csv.png", dpi=300)
plt.show()
```
