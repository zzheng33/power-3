```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
# Define the layers of the MLP split into two parallel columns
layers_left = [
    ("Input", 1),
    ("Dense (256, ReLU)", 256),
    ("BatchNorm", 256),
    ("Dropout (0.3)", 256),
    ("Dense (128, ReLU)", 128)
]

layers_right = [
    ("BatchNorm", 128),
    ("Dropout (0.3)", 128),
    ("Dense (64, SELU)", 64),
    ("Dense (32, SELU)", 32),
    ("Output (1, Linear)", 1)
]

# Create a directed graph
G = nx.DiGraph()

# Add nodes for each layer
for i, (layer_name, size) in enumerate(layers_left):
    G.add_node(f"L{i}", label=layer_name, size=size)
for i, (layer_name, size) in enumerate(layers_right):
    G.add_node(f"R{i}", label=layer_name, size=size)

# Add edges to represent connections between layers (parallel branches)
for i in range(len(layers_left) - 1):
    G.add_edge(f"L{i}", f"L{i+1}")
for i in range(len(layers_right) - 1):
    G.add_edge(f"R{i}", f"R{i+1}")

# Connect the last node of the left column to the first node of the right column in parallel
G.add_edge(f"L{len(layers_left)-1}", "R0")

# Define positions for a **closer** two-column layout
pos = {}
x_offset = 0.002  # Reduce space between branches
for i in range(len(layers_left)):
    pos[f"L{i}"] = (-x_offset, -i)  # Left column (aligned vertically)
    pos[f"R{i}"] = (x_offset, -i)  # Right column (same height as left column)

# Create labels with layer names and sizes
labels = {node: f"{G.nodes[node]['label']}" for node in G.nodes()}

# Draw the graph with aligned parallel branches, larger circles, and bigger text
plt.figure(figsize=(7, 5))
nx.draw(G, pos, with_labels=True, labels=labels, node_size=3000,  # Larger circles
        node_color="lightblue", edge_color="black", font_size=10,  # Bigger text
        font_weight="bold", arrows=True, arrowsize=12, width=1.5)
# plt.title("MLP Architecture (Closer Parallel Branches)", fontsize=14, fontweight="bold")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define architecture layers
gate_layers = [
    ("Input (8 features)", 8),
    ("Dense (64, ReLU)", 64),
    ("Dense (32, ReLU)", 32),
    ("Dense (16, ReLU)", 16),
    ("Dense (8, ReLU)", 8),
    ("Sigmoid Output (Gate)", 1)
]

cpu_layers = [
    ("Dense (256, ReLU)", 256),
    ("Dense (128, ReLU)", 128),
    ("Dense (64, SELU)", 64),
    ("Dense (32, SELU)", 32),
    ("Dense (1)", 1)
]

gpu_layers = list(cpu_layers)

# Create graph
G = nx.DiGraph()
gate_nodes, cpu_nodes, gpu_nodes = [], [], []

# Add nodes
for i, (name, _) in enumerate(gate_layers):
    node = f"G{i}"
    G.add_node(node, label=name)
    gate_nodes.append(node)

for i, (name, _) in enumerate(cpu_layers):
    node = f"C{i}"
    G.add_node(node, label=name)
    cpu_nodes.append(node)

for i, (name, _) in enumerate(gpu_layers):
    node = f"U{i}"
    G.add_node(node, label=name)
    gpu_nodes.append(node)

# Add fusion node
fusion_node = "F"
G.add_node(fusion_node, label="Fusion\nWeighted Sum")

# Define edges
gate_edges = [(gate_nodes[i], gate_nodes[i + 1]) for i in range(len(gate_nodes) - 1)]
cpu_edges = [(cpu_nodes[i], cpu_nodes[i + 1]) for i in range(len(cpu_nodes) - 1)]
gpu_edges = [(gpu_nodes[i], gpu_nodes[i + 1]) for i in range(len(gpu_nodes) - 1)]
fusion_edges = [("G5", fusion_node), ("C4", fusion_node), ("U4", fusion_node)]

G.add_edges_from(gate_edges + cpu_edges + gpu_edges + fusion_edges)

# Positioning
pos = {}
offset = 0.01
for i, node in enumerate(gate_nodes):
    pos[node] = (0, -i)
for i, node in enumerate(cpu_nodes):
    pos[node] = (-offset, -i)
for i, node in enumerate(gpu_nodes):
    pos[node] = (offset, -i)
pos[fusion_node] = (0, -len(gate_nodes))

# Labels and colors
labels = {n: G.nodes[n]['label'] for n in G.nodes()}
node_colors = ["orange" if n.startswith("G") else "skyblue" if n.startswith("C") else
               "lightgreen" if n.startswith("U") else "lightgray" for n in G.nodes()]
node_size = 3000

# Draw
plt.figure(figsize=(8, 8))
nx.draw(G, pos, with_labels=True, labels=labels, node_size=node_size,
        node_color=node_colors, edge_color="black", font_size=10,
        font_weight="bold", arrows=True, arrowsize=20, connectionstyle="arc3,rad=0.0")

# Section labels
plt.text(-offset, 0.5, "CPU Expert", ha='center', fontsize=12, fontweight='bold', color='blue')
plt.text(0, 0.5, "Soft Gating Network", ha='center', fontsize=12, fontweight='bold', color='orange')
plt.text(offset, 0.5, "GPU Expert", ha='center', fontsize=12, fontweight='bold', color='green')

plt.axis("off")
plt.savefig("./fig/MLP.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Define figure size
fig, ax = plt.subplots(figsize=(10, 10))

# Define positions for elements with layer details
positions = {
    "Input Power Pair\n(1)": (2, 10),
    "Embedding (Power Pair)\n(latent_dim)": (2, 9),
    "Flatten (Power Pair)\n(latent_dim)": (2, 8),

    "Input Application\n(1)": (6, 10),
    "Embedding (Application)\n(latent_dim)": (6, 9),
    "Flatten (Application)\n(latent_dim)": (6, 8),

    "Concatenate\n(2 * latent_dim)": (4, 7),

    "Dense (128, SELU)\n(128 neurons)": (4, 6),
    "Dense (64, SELU)\n(64 neurons)": (4, 5),
    "Dense (32, SELU)\n(32 neurons)": (4, 4),
    "Output (1, Linear)\n(1 neuron)": (4, 3)
}

# Define arrows (connections between layers)
connections = [
    ("Input Power Pair\n(1)", "Embedding (Power Pair)\n(latent_dim)"),
    ("Embedding (Power Pair)\n(latent_dim)", "Flatten (Power Pair)\n(latent_dim)"),

    ("Input Application\n(1)", "Embedding (Application)\n(latent_dim)"),
    ("Embedding (Application)\n(latent_dim)", "Flatten (Application)\n(latent_dim)"),

    ("Flatten (Power Pair)\n(latent_dim)", "Concatenate\n(2 * latent_dim)"),
    ("Flatten (Application)\n(latent_dim)", "Concatenate\n(2 * latent_dim)"),

    ("Concatenate\n(2 * latent_dim)", "Dense (128, SELU)\n(128 neurons)"),
    ("Dense (128, SELU)\n(128 neurons)", "Dense (64, SELU)\n(64 neurons)"),
    ("Dense (64, SELU)\n(64 neurons)", "Dense (32, SELU)\n(32 neurons)"),
    ("Dense (32, SELU)\n(32 neurons)", "Output (1, Linear)\n(1 neuron)")
]

# Define light colors for different layers
layer_colors = {
    "Input Layer": "#FFECB3",  # Light Yellow
    "Embedding Layer": "#FFCCBC",  # Light Red-Orange
    "Concatenation Layer": "#C8E6C9",  # Light Green
    "Neural CF Layers": "#BBDEFB",  # Light Blue
    "Output Layer": "#E1BEE7"  # Light Purple
}

# Assign colors to each layer type
layer_color_map = {
    "Input Power Pair\n(1)": layer_colors["Input Layer"],
    "Embedding (Power Pair)\n(latent_dim)": layer_colors["Embedding Layer"],
    "Flatten (Power Pair)\n(latent_dim)": layer_colors["Embedding Layer"],

    "Input Application\n(1)": layer_colors["Input Layer"],
    "Embedding (Application)\n(latent_dim)": layer_colors["Embedding Layer"],
    "Flatten (Application)\n(latent_dim)": layer_colors["Embedding Layer"],

    "Concatenate\n(2 * latent_dim)": layer_colors["Concatenation Layer"],

    "Dense (128, SELU)\n(128 neurons)": layer_colors["Neural CF Layers"],
    "Dense (64, SELU)\n(64 neurons)": layer_colors["Neural CF Layers"],
    "Dense (32, SELU)\n(32 neurons)": layer_colors["Neural CF Layers"],
    "Output (1, Linear)\n(1 neuron)": layer_colors["Output Layer"]
}

# Draw elements as rounded rectangles with light colors
for label, (x, y) in positions.items():
    rect = FancyBboxPatch((x - 1, y - 0.3), 2, 0.6, boxstyle="round,pad=0.2",
                           edgecolor="black", facecolor=layer_color_map[label])
    ax.add_patch(rect)
    ax.text(x, y, label, ha="center", va="center", fontsize=10, fontweight="bold", color="black")

# Draw arrows for connections
for start, end in connections:
    start_x, start_y = positions[start]
    end_x, end_y = positions[end]
    ax.annotate("", xy=(end_x, end_y + 0.3), xytext=(start_x, start_y - 0.3),
                arrowprops=dict(arrowstyle="->", color="black", lw=1.5))

# Add side text labels for different sections
text_positions = {
    "Input Layer (Sparse)": (8, 10),
    "Embedding Layer": (8, 9),
    "Neural CF Layers": (8, 6),
    "Output Layer": (8, 3)
}

for text, (x, y) in text_positions.items():
    ax.text(x, y, text, ha="left", va="center", fontsize=12, fontweight="bold")

# Adjust plot limits and remove axes
ax.set_xlim(0, 10)
ax.set_ylim(2, 11)
ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")

# Title
# plt.title("Neural Collaborative Filtering (NCF) Model", fontsize=12, fontweight="bold")

# Show plot
plt.show()
```
