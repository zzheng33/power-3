```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.decomposition import TruncatedSVD
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dot, Dense, Concatenate,LayerNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.impute import SimpleImputer
from tensorflow.keras.layers import Add, LeakyReLU
from scipy import stats
from tensorflow.keras.activations import relu
seed = 1
```

```python
############################# Step 1: Data Preprocessing ############################# 

# Define directories for training and validation data
train_dir = "./altis_power_cap_res/runs/train"
validation_dir = "./ecp_power_cap_res/runs/validation"
# validation_dir = "./hec_power_cap_res/runs/validation"
# validation_dir = "./altis_power_cap_res/runs/validation"

# train_dir = "./npb_power_cap_res/runs/train"
# validation_dir = "./npb_power_cap_res/runs/validation"

save_model_dir = "./MLP_model/gpu_model/arxiv/"
os.makedirs(save_model_dir, exist_ok=True)

# Load CSV files from train directory
train_csv_files = [f for f in os.listdir(train_dir) if f.endswith(".csv")]
validation_csv_files = [f for f in os.listdir(validation_dir) if f.endswith(".csv")]

# Load and merge training data while filtering out specified CPU power caps
train_data = []
for file in train_csv_files:
    file_path = os.path.join(train_dir, file)
    df = pd.read_csv(file_path)

    # Drop rows where CPU Power Cap is 120 or 130
    df = df[~df["CPU Power Cap"].isin([120, 130])]

    train_data.append(df)

# Merge all training data
train_df = pd.concat(train_data, ignore_index=True)

# Load validation data while filtering out specified CPU power caps
validation_data = {}
for file in validation_csv_files:
    file_path = os.path.join(validation_dir, file)
    df = pd.read_csv(file_path)

    # Drop unwanted GPU Power Cap values
    df = df[~df["GPU Power Cap"].isin([123])]

    # Reset index to maintain row-order alignment
    df = df.sort_values(by=["CPU Power Cap", "GPU Power Cap"]).reset_index(drop=True)

    validation_data[file] = df


# Drop any rows with missing values
train_df.dropna(inplace=True)
for key in validation_data:
    validation_data[key].dropna(inplace=True)
```

```python
############################# Complete Model Comparison with MLP #############################
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import seaborn as sns

# Advanced gradient boosting frameworks
try:
    import xgboost as xgb
    XGBOOST_AVAILABLE = True
except ImportError:
    XGBOOST_AVAILABLE = False
    print("XGBoost not available - install with: pip install xgboost")

try:
    import lightgbm as lgb
    LIGHTGBM_AVAILABLE = True
except ImportError:
    LIGHTGBM_AVAILABLE = False
    print("LightGBM not available - install with: pip install lightgbm")

# For TensorFlow/Keras MLP (if available)
try:
    import tensorflow as tf
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
    from tensorflow.keras.optimizers import Adam
    from tensorflow.keras.callbacks import EarlyStopping
    TENSORFLOW_AVAILABLE = True
except ImportError:
    TENSORFLOW_AVAILABLE = False
    print("TensorFlow not available, will skip Keras MLP comparison")

# Define feature columns and target column (using your exact setup)
feature_cols = ["CPU Power Cap", "GPU Power Cap", "Memory Throughput", "SM Clock", "DRAM Active", "FP Active"]
target_col = "Performance"

# Extract features and target for training (matching your exact setup)
X_train = train_df[feature_cols].values
y_train = train_df[target_col].values.reshape(-1, 1)

# Normalize features and target (using your exact scalers)
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
y_train_scaled = scaler_y.fit_transform(y_train)

# For scikit-learn models, we need flattened y_train
y_train_flat = y_train.flatten()

############################# Model 1: Linear Regression #############################
print("="*50)
print("LINEAR REGRESSION")
print("="*50)

# Simple Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train_flat)

# Predictions
y_pred_linear = linear_model.predict(X_train_scaled)

# Metrics
mse_linear = mean_squared_error(y_train_flat, y_pred_linear)
r2_linear = r2_score(y_train_flat, y_pred_linear)
mae_linear = mean_absolute_error(y_train_flat, y_pred_linear)

print(f"Linear Regression Results:")
print(f"MSE: {mse_linear:.4f}")
print(f"R²: {r2_linear:.4f}")
print(f"MAE: {mae_linear:.4f}")

# Feature importance (coefficients)
feature_importance_linear = pd.DataFrame({
    'Feature': feature_cols,
    'Coefficient': linear_model.coef_,
    'Abs_Coefficient': np.abs(linear_model.coef_)
}).sort_values('Abs_Coefficient', ascending=False)
print("\nFeature Importance (Linear Regression):")
print(feature_importance_linear)

############################# Model 2: Polynomial Regression #############################
print("\n" + "="*50)
print("POLYNOMIAL REGRESSION")
print("="*50)

# Polynomial Regression with degree 2 and 3
for degree in [2, 3]:
    print(f"\nPolynomial Regression (degree {degree}):")
    
    # Create polynomial features
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly_features.fit_transform(X_train_scaled)
    
    # Fit linear regression on polynomial features
    poly_model = LinearRegression()
    poly_model.fit(X_poly, y_train_flat)
    
    # Predictions
    y_pred_poly = poly_model.predict(X_poly)
    
    # Metrics
    mse_poly = mean_squared_error(y_train_flat, y_pred_poly)
    r2_poly = r2_score(y_train_flat, y_pred_poly)
    mae_poly = mean_absolute_error(y_train_flat, y_pred_poly)
    
    print(f"MSE: {mse_poly:.4f}")
    print(f"R²: {r2_poly:.4f}")
    print(f"MAE: {mae_poly:.4f}")

############################# Model 3: Regularized Linear Models #############################
print("\n" + "="*50)
print("REGULARIZED LINEAR MODELS")
print("="*50)

# Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled, y_train_flat)
y_pred_ridge = ridge_model.predict(X_train_scaled)
print(f"Ridge Regression - MSE: {mean_squared_error(y_train_flat, y_pred_ridge):.4f}, R²: {r2_score(y_train_flat, y_pred_ridge):.4f}")

# Note: Lasso and SVR removed as requested

############################# Model 4: Tree-Based Models #############################
print("\n" + "="*50)
print("TREE-BASED MODELS")
print("="*50)

# Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train_flat)
y_pred_rf = rf_model.predict(X_train_scaled)
print(f"Random Forest - MSE: {mean_squared_error(y_train_flat, y_pred_rf):.4f}, R²: {r2_score(y_train_flat, y_pred_rf):.4f}")

# Feature importance from Random Forest
feature_importance_rf = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)
print("Random Forest Feature Importance:")
print(feature_importance_rf)

# Gradient Boosting
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train_scaled, y_train_flat)
y_pred_gb = gb_model.predict(X_train_scaled)
print(f"\nGradient Boosting - MSE: {mean_squared_error(y_train_flat, y_pred_gb):.4f}, R²: {r2_score(y_train_flat, y_pred_gb):.4f}")

# Decision Tree (single tree)
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train_scaled, y_train_flat)
y_pred_dt = dt_model.predict(X_train_scaled)
print(f"Decision Tree - MSE: {mean_squared_error(y_train_flat, y_pred_dt):.4f}, R²: {r2_score(y_train_flat, y_pred_dt):.4f}")

############################# Model 5: Advanced Gradient Boosting (XGBoost & LightGBM) #############################
print("\n" + "="*50)
print("ADVANCED GRADIENT BOOSTING")
print("="*50)

# XGBoost
if XGBOOST_AVAILABLE:
    xgb_model = xgb.XGBRegressor(
        n_estimators=100,
        max_depth=6,
        learning_rate=0.1,
        random_state=42,
        eval_metric='rmse'
    )
    xgb_model.fit(X_train_scaled, y_train_flat)
    y_pred_xgb = xgb_model.predict(X_train_scaled)
    print(f"XGBoost - MSE: {mean_squared_error(y_train_flat, y_pred_xgb):.4f}, R²: {r2_score(y_train_flat, y_pred_xgb):.4f}")
    
    # Feature importance from XGBoost
    feature_importance_xgb = pd.DataFrame({
        'Feature': feature_cols,
        'Importance': xgb_model.feature_importances_
    }).sort_values('Importance', ascending=False)
    print("XGBoost Feature Importance:")
    print(feature_importance_xgb)
else:
    print("XGBoost not available")
    xgb_model = None

# LightGBM
if LIGHTGBM_AVAILABLE:
    lgb_model = lgb.LGBMRegressor(
        n_estimators=100,
        max_depth=6,
        learning_rate=0.1,
        random_state=42,
        verbose=-1  # Suppress warnings
    )
    lgb_model.fit(X_train_scaled, y_train_flat)
    y_pred_lgb = lgb_model.predict(X_train_scaled)
    print(f"\nLightGBM - MSE: {mean_squared_error(y_train_flat, y_pred_lgb):.4f}, R²: {r2_score(y_train_flat, y_pred_lgb):.4f}")
    
    # Feature importance from LightGBM
    feature_importance_lgb = pd.DataFrame({
        'Feature': feature_cols,
        'Importance': lgb_model.feature_importances_
    }).sort_values('Importance', ascending=False)
    print("LightGBM Feature Importance:")
    print(feature_importance_lgb)
else:
    print("LightGBM not available")
    lgb_model = None

############################# Model 6: K-Nearest Neighbors #############################
print("\n" + "="*50)
print("K-NEAREST NEIGHBORS")
print("="*50)

# KNN with different k values
k_values = [3, 5, 7, 10]
for k in k_values:
    knn_model = KNeighborsRegressor(n_neighbors=k)
    knn_model.fit(X_train_scaled, y_train_flat)
    y_pred_knn = knn_model.predict(X_train_scaled)
    print(f"KNN (k={k}) - MSE: {mean_squared_error(y_train_flat, y_pred_knn):.4f}, R²: {r2_score(y_train_flat, y_pred_knn):.4f}")

############################# Model 7: Multi-Layer Perceptron #############################
if TENSORFLOW_AVAILABLE:
    print("\n" + "="*50)
    print("MULTI-LAYER PERCEPTRON (TENSORFLOW/KERAS)")
    print("="*50)
    
    # Build your exact MLP model architecture
    model = Sequential([
        tf.keras.Input(shape=(X_train_scaled.shape[1],)),  # Explicitly define input shape
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.1),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.1),
        Dense(64, activation='relu'),
        Dense(32, activation='selu'),
        Dense(1, activation='linear')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )
    
    # Early stopping callback
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=50,
        restore_best_weights=True
    )
    
    print("Training MLP model...")
    history = model.fit(
        X_train_scaled, y_train_scaled,
        epochs=500,
        batch_size=32,
        validation_split=0.2,
        callbacks=[early_stopping],
        verbose=0  # Set to 1 if you want to see training progress
    )
    
    # Make predictions and inverse transform to original scale
    y_pred_keras_scaled = model.predict(X_train_scaled, verbose=0)
    y_pred_keras = scaler_y.inverse_transform(y_pred_keras_scaled).flatten()
    
    print(f"MLP - MSE: {mean_squared_error(y_train_flat, y_pred_keras):.4f}, R²: {r2_score(y_train_flat, y_pred_keras):.4f}")
    print(f"Training stopped at epoch: {len(history.history['loss'])}")
    
    # Plot training history
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('MLP Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Training MAE')
    plt.plot(history.history['val_mae'], label='Validation MAE')
    plt.title('MLP Training MAE')
    plt.xlabel('Epoch')
    plt.ylabel('MAE')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Store the model for later use
    keras_mlp = model
    
else:
    print("\nMLP skipped - TensorFlow not available")
    keras_mlp = None
    y_pred_keras = None

############################# Model Comparison and Selection #############################
print("\n" + "="*50)
print("MODEL COMPARISON WITH CROSS-VALIDATION")
print("="*50)

# Create a dictionary of models for comparison (including XGBoost and LightGBM)
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'KNN (k=5)': KNeighborsRegressor(n_neighbors=5)
}

# Add XGBoost and LightGBM if available
if XGBOOST_AVAILABLE:
    models['XGBoost'] = xgb.XGBRegressor(
        n_estimators=100,
        max_depth=6,
        learning_rate=0.1,
        random_state=42,
        eval_metric='rmse'
    )

if LIGHTGBM_AVAILABLE:
    models['LightGBM'] = lgb.LGBMRegressor(
        n_estimators=100,
        max_depth=6,
        learning_rate=0.1,
        random_state=42,
        verbose=-1
    )

# Perform cross-validation
cv_results = {}
for name, model in models.items():
    print(f"Cross-validating {name}...")
    cv_scores = cross_val_score(model, X_train_scaled, y_train_flat, cv=5, scoring='r2')
    cv_results[name] = {
        'CV_Mean_R2': cv_scores.mean(),
        'CV_Std_R2': cv_scores.std()
    }
    print(f"{name}: R² = {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")

# Add MLP results if available
if TENSORFLOW_AVAILABLE and keras_mlp is not None:
    # For MLP, we'll use the training R² as an approximation
    # since cross-validation with Keras is more complex
    keras_r2 = r2_score(y_train_flat, y_pred_keras)
    cv_results['MLP'] = {
        'CV_Mean_R2': keras_r2,  # Using training R² as approximation
        'CV_Std_R2': 0.0  # Not available for single training
    }
    print(f"MLP: R² = {keras_r2:.4f} (training score)")

# Create a summary comparison table
print("\n" + "="*70)
print("MODEL COMPARISON SUMMARY")
print("="*70)
comparison_df = pd.DataFrame(cv_results).T
comparison_df = comparison_df.sort_values('CV_Mean_R2', ascending=False)
comparison_df['CV_Mean_R2'] = comparison_df['CV_Mean_R2'].round(4)
comparison_df['CV_Std_R2'] = comparison_df['CV_Std_R2'].round(4)
print(comparison_df)

# Create enhanced model comparison visualizations
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Model Performance Comparison Dashboard', fontsize=16, fontweight='bold', y=0.98)

# Color scheme
colors = plt.cm.viridis(np.linspace(0.2, 0.9, len(comparison_df)))
highlight_color = '#ff6b35'  # Orange for best model

model_names = list(comparison_df.index)
mean_scores = comparison_df['CV_Mean_R2'].values
std_scores = comparison_df['CV_Std_R2'].values

# Plot 1: Horizontal bar chart with error bars
bar_colors = [highlight_color if score == max(mean_scores) else colors[i] for i, score in enumerate(mean_scores)]
bars = ax1.barh(range(len(model_names)), mean_scores, xerr=std_scores, 
                capsize=4, color=bar_colors, alpha=0.8, edgecolor='white', linewidth=1.5)

ax1.set_yticks(range(len(model_names)))
ax1.set_yticklabels(model_names, fontsize=10)
ax1.set_xlabel('Cross-Validation R² Score', fontsize=12, fontweight='bold')
ax1.set_title('Model Performance with Confidence Intervals', fontsize=14, fontweight='bold', pad=20)
ax1.grid(axis='x', alpha=0.3, linestyle='--')
ax1.set_xlim(0, max(mean_scores) * 1.15)

# Add value labels on bars
for i, (mean, std) in enumerate(zip(mean_scores, std_scores)):
    ax1.text(mean + max(mean_scores) * 0.02, i, f'{mean:.3f}', 
             va='center', fontweight='bold', fontsize=10)

# Add best model annotation
best_idx = np.argmax(mean_scores)
ax1.annotate('🏆 Best Model', xy=(mean_scores[best_idx], best_idx), 
             xytext=(mean_scores[best_idx] + max(mean_scores) * 0.1, best_idx),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=11, fontweight='bold', color='red')

# Plot 2: Vertical bar chart (clean version)
bars2 = ax2.bar(range(len(model_names)), mean_scores, color=bar_colors, alpha=0.8, 
                edgecolor='white', linewidth=1.5)
ax2.set_xticks(range(len(model_names)))
ax2.set_xticklabels([name.replace(' ', '\n') for name in model_names], 
                    rotation=0, fontsize=9, ha='center')
ax2.set_ylabel('Cross-Validation R² Score', fontsize=12, fontweight='bold')
ax2.set_title('Model Performance Overview', fontsize=14, fontweight='bold', pad=20)
ax2.grid(axis='y', alpha=0.3, linestyle='--')
ax2.set_ylim(0, max(mean_scores) * 1.1)

# Add value labels on top of bars
for i, mean in enumerate(mean_scores):
    ax2.text(i, mean + max(mean_scores) * 0.02, f'{mean:.3f}', 
             ha='center', va='bottom', fontweight='bold', fontsize=10)

# Plot 3: Performance distribution (violin plot style)
positions = range(len(model_names))
parts = ax3.violinplot([np.random.normal(mean, std, 100) for mean, std in zip(mean_scores, std_scores)], 
                       positions=positions, widths=0.7, showmeans=True, showmedians=True)

for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(bar_colors[i])
    pc.set_alpha(0.7)

ax3.set_xticks(positions)
ax3.set_xticklabels([name.replace(' ', '\n') for name in model_names], 
                    rotation=0, fontsize=9, ha='center')
ax3.set_ylabel('R² Score Distribution', fontsize=12, fontweight='bold')
ax3.set_title('Performance Variability Analysis', fontsize=14, fontweight='bold', pad=20)
ax3.grid(axis='y', alpha=0.3, linestyle='--')

# Plot 4: Model ranking with performance gaps
sorted_indices = np.argsort(mean_scores)[::-1]
sorted_names = [model_names[i] for i in sorted_indices]
sorted_scores = [mean_scores[i] for i in sorted_indices]
sorted_colors = [bar_colors[i] for i in sorted_indices]

y_pos = range(len(sorted_names))
bars4 = ax4.barh(y_pos, sorted_scores, color=sorted_colors, alpha=0.8, 
                 edgecolor='white', linewidth=1.5)

ax4.set_yticks(y_pos)
ax4.set_yticklabels(sorted_names, fontsize=10)
ax4.set_xlabel('Cross-Validation R² Score', fontsize=12, fontweight='bold')
ax4.set_title('Final Model Ranking', fontsize=14, fontweight='bold', pad=20)
ax4.grid(axis='x', alpha=0.3, linestyle='--')

# Add ranking numbers and performance gaps
for i, (score, name) in enumerate(zip(sorted_scores, sorted_names)):
    # Ranking number
    ax4.text(-max(sorted_scores) * 0.05, i, f'#{i+1}', 
             ha='center', va='center', fontweight='bold', fontsize=12, 
             bbox=dict(boxstyle='circle', facecolor='white', edgecolor='gray'))
    
    # Score label
    ax4.text(score + max(sorted_scores) * 0.02, i, f'{score:.3f}', 
             va='center', fontweight='bold', fontsize=10)
    
    # Performance gap (except for best model)
    if i > 0:
        gap = sorted_scores[0] - score
        ax4.text(score + max(sorted_scores) * 0.08, i, f'(-{gap:.3f})', 
                 va='center', fontsize=9, color='red', style='italic')

ax4.set_xlim(-max(sorted_scores) * 0.1, max(sorted_scores) * 1.2)

# Add a subtle background gradient to the entire figure
fig.patch.set_facecolor('#f8f9fa')

plt.tight_layout()
plt.subplots_adjust(top=0.93, hspace=0.3, wspace=0.3)
plt.show()

# Create a summary statistics table plot
fig2, ax_table = plt.subplots(figsize=(12, 6))
ax_table.axis('tight')
ax_table.axis('off')

# Prepare data for the table
table_data = []
for i, (name, row) in enumerate(comparison_df.iterrows()):
    rank = sorted_indices.tolist().index(i) + 1
    performance_tier = "🥇 Excellent" if rank <= 2 else "🥈 Good" if rank <= 4 else "🥉 Fair"
    table_data.append([
        f"#{rank}",
        name,
        f"{row['CV_Mean_R2']:.4f}",
        f"±{row['CV_Std_R2']:.4f}",
        performance_tier
    ])

# Sort by rank
table_data.sort(key=lambda x: int(x[0][1:]))

# Create table
table = ax_table.table(cellText=table_data,
                      colLabels=['Rank', 'Model', 'Mean R²', 'Std Dev', 'Performance Tier'],
                      cellLoc='center',
                      loc='center',
                      colWidths=[0.1, 0.35, 0.15, 0.15, 0.25])

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 2)

# Color code the table
for i in range(len(table_data)):
    for j in range(5):
        cell = table[(i+1, j)]
        if i == 0:  # Best model
            cell.set_facecolor('#e8f5e8')
            cell.set_text_props(weight='bold')
        elif i < 3:  # Top 3
            cell.set_facecolor('#f0f8ff')
        else:
            cell.set_facecolor('#f9f9f9')

# Header styling
for j in range(5):
    table[(0, j)].set_facecolor('#34495e')
    table[(0, j)].set_text_props(weight='bold', color='white')

plt.title('Model Performance Summary Table', fontsize=16, fontweight='bold', pad=20)
plt.show()

############################# Hyperparameter Tuning for Best Model #############################
print("\n" + "="*50)
print("HYPERPARAMETER TUNING")
print("="*50)

# Find best model for tuning (excluding MLP for grid search)
traditional_models = {k: v for k, v in cv_results.items() if k != 'MLP'}
best_traditional_name = max(traditional_models, key=lambda x: traditional_models[x]['CV_Mean_R2'])

print(f"Tuning hyperparameters for: {best_traditional_name}")

if best_traditional_name == 'Random Forest':
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }
    base_model = RandomForestRegressor(random_state=42)
elif best_traditional_name == 'XGBoost' and XGBOOST_AVAILABLE:
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 6, 9],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.8, 1.0]
    }
    base_model = xgb.XGBRegressor(random_state=42, eval_metric='rmse')
elif best_traditional_name == 'LightGBM' and LIGHTGBM_AVAILABLE:
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 6, 9],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.8, 1.0]
    }
    base_model = lgb.LGBMRegressor(random_state=42, verbose=-1)
else:
    # Default to Random Forest
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20]
    }
    base_model = RandomForestRegressor(random_state=42)

grid_search = GridSearchCV(
    base_model,
    param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

grid_search.fit(X_train_scaled, y_train_flat)
print(f"Best {best_traditional_name} Parameters:")
print(grid_search.best_params_)
print(f"Best Cross-Validation R²: {grid_search.best_score_:.4f}")

# Train final model with best parameters
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_train_scaled)
print(f"Best {best_traditional_name} - Training MSE: {mean_squared_error(y_train_flat, y_pred_best):.4f}")
print(f"Best {best_traditional_name} - Training R²: {r2_score(y_train_flat, y_pred_best):.4f}")

############################# Final Model Recommendations #############################
print("\n" + "="*50)
print("FINAL RECOMMENDATIONS")
print("="*50)

# Find the best performing model
best_model_name = max(cv_results, key=lambda x: cv_results[x]['CV_Mean_R2'])
print(f"🏆 Best performing model: {best_model_name}")
print(f"   Cross-validation R²: {cv_results[best_model_name]['CV_Mean_R2']:.4f}")
print(f"   Standard deviation: {cv_results[best_model_name]['CV_Std_R2']:.4f}")

# Show top 3 models
top_3_models = sorted(cv_results.items(), key=lambda x: x[1]['CV_Mean_R2'], reverse=True)[:3]
print(f"\n🥇 Top 3 Models:")
for i, (name, scores) in enumerate(top_3_models, 1):
    print(f"   {i}. {name}: R² = {scores['CV_Mean_R2']:.4f} (±{scores['CV_Std_R2']:.4f})")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Create the simplified neural network topology diagram
fig, ax = plt.subplots(figsize=(11, 8))  # Made narrower to accommodate closer layers
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.axis('off')

# Define layer positions and sizes - CLOSER SPACING
layer_x_positions = [1.5, 3, 4.5, 6, 7.5, 8.5]  # Reduced spacing between layers
layer_names = ['Input', 'Dense\n256', 'Dense\n128', 'Dense\n64', 'Dense\n32', 'Output']
layer_sizes = [7, 7, 4, 3, 2, 1]  # Input: 7, Dense 256: 7, Dense 128: 4, etc.

# Function to draw a neuron
def draw_neuron(x, y, radius=0.2, color='white'):  # Increased radius from 0.12 to 0.2
    circle = plt.Circle((x, y), radius, facecolor=color, edgecolor='black', linewidth=1.5)  # Slightly thicker border
    ax.add_patch(circle)

# Function to draw connections between layers (simplified)
def draw_connections(x1, y1_list, x2, y2_list, alpha=0.4):
    # For input layer to first hidden layer, show all connections
    if x1 == layer_x_positions[0]:  # Input layer connections
        # Show connections from ALL input neurons to ALL visible neurons in next layer
        for y1 in y1_list:  # All 7 input neurons
            for y2 in y2_list:  # All 7 visible neurons in Dense 256
                ax.plot([x1, x2], [y1, y2], 'k-', alpha=alpha, linewidth=0.5)
    else:
        # For other layers, use sparse connections to avoid clutter
        sample_y1 = y1_list[::max(1, len(y1_list)//3)]  # Sample every few neurons
        sample_y2 = y2_list[::max(1, len(y2_list)//3)]
        
        for y1 in sample_y1:
            for y2 in sample_y2:
                ax.plot([x1, x2], [y1, y2], 'k-', alpha=alpha, linewidth=0.5)

# Draw neurons for each layer
layer_positions = []
for i, (x, size) in enumerate(zip(layer_x_positions, layer_sizes)):
    if i == len(layer_x_positions) - 1:  # Output layer - center it
        y_positions = [4]  # Center position (middle of 2-6 range)
    else:
        y_positions = np.linspace(2, 6, size)
    layer_positions.append(y_positions)
    
    # Determine color
    if i == 0:  # Input layer
        color = 'lightblue'
    elif i == len(layer_x_positions) - 1:  # Output layer
        color = 'lightgreen'
    else:  # Hidden layers
        color = 'lightyellow'
    
    # Draw neurons
    for y in y_positions:
        draw_neuron(x, y, color=color)
    
    # Add dots for hidden layers to indicate more neurons
    if i in [1, 2, 3, 4]:  # All hidden layers (256, 128, 64, 32)
        ax.text(x, 1.5, '⋮', ha='center', va='center', fontsize=16, fontweight='bold')
        ax.text(x, 6.5, '⋮', ha='center', va='center', fontsize=16, fontweight='bold')

# Draw simplified connections between layers
for i in range(len(layer_positions) - 1):
    x1, x2 = layer_x_positions[i], layer_x_positions[i + 1]
    y1_list, y2_list = layer_positions[i], layer_positions[i + 1]
    
    # Special handling for input to first hidden layer - show ALL connections
    if i == 0:  # Input to Dense 256
        for y1 in y1_list:  # Every input neuron
            for y2 in y2_list:  # Every visible neuron in Dense 256
                ax.plot([x1 + 0.2, x2 - 0.2], [y1, y2], 'k-', alpha=0.4, linewidth=0.5)
    else:
        # For other layers, use the function
        draw_connections(x1 + 0.2, y1_list, x2 - 0.2, y2_list)

# Add layer labels
label_y = 0.7
for i, (x, name) in enumerate(zip(layer_x_positions, layer_names)):
    ax.text(x, label_y, name, ha='center', va='center', fontsize=18, fontweight='bold')

# Add input labels (including CPU and GPU Power Cap)
input_features = [
    'cpuPowerCap',
    'gpuPowerCap',
    'Mem. Tht',
    'SM Clock',
    'DRAM Active',
    'FP Active',
    'IPS'
]

for i, (feature, y) in enumerate(zip(input_features, layer_positions[0])):
    ax.text(-1, y, feature, ha='left', va='center', fontsize=18, fontweight='bold')
    # Draw arrow from label to input neuron
    ax.annotate('', xy=(layer_x_positions[0] - 0.2, y), xytext=(1.2, y),  # Adjusted for closer spacing
                arrowprops=dict(arrowstyle='->', lw=1, color='black'))

# Add simplified activation labels - adjusted positions for closer layers
ax.text(3.75, 7.2, 'ReLU + BatchNorm + Dropout', ha='center', va='center', fontsize=16, 
        fontweight='bold')
ax.text(6, 7.2, 'ReLU', ha='center', va='center', fontsize=16, 
        fontweight='bold')
ax.text(7.5, 7.2, 'SELU', ha='center', va='center', fontsize=16, 
        fontweight='bold')

plt.tight_layout()

# Save the diagram
plt.savefig('./fig/mlp_topology.png', dpi=300, bbox_inches='tight', 
            facecolor='white', edgecolor='none')
plt.show()

print("Simplified MLP Topology Diagram with closer layers saved as:")
print("  - ./fig/mlp_topology.png")
```
