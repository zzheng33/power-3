### Vioin plot for prediction accuracy distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def read_app_error_dict_ordered(csv_path):
    df = pd.read_csv(csv_path)

    # Get application order as they first appear in the CSV
    app_order = df["Application"].drop_duplicates().tolist()

    # Group manually to preserve order
    app_error_dict_cf = {
        app: df[df["Application"] == app]["Prediction Accuracy (%)"].tolist()
        for app in app_order
    }

    return app_error_dict_cf
    
def plot_ncf_acc_violin(csv_path, x_rows=3, y_cols=4, y_lim=(70, 100)):
    # Read CSV
    df = pd.read_csv(csv_path)

    # Group into dictionary: app â†’ list of accuracy values
    app_error_dict_cf = read_app_error_dict_ordered(csv_path)

    

    # Create subplot grid
    num_apps = len(app_error_dict_cf)
    fig, axes = plt.subplots(x_rows, y_cols, figsize=(y_cols * 3, x_rows * 3), sharey=True)
    axes = axes.flatten()

    for idx, (app, errors) in enumerate(app_error_dict_cf.items()):
        accuracies = [100 - e for e in errors]
        df_plot = pd.DataFrame({"Prediction Accuracy (%)": accuracies})
        ax = axes[idx]
        # print(df_plot)
        sns.violinplot(
            y="Prediction Accuracy (%)",
            data=df_plot,
            cut=0,
            bw_method=0.3,
            ax=ax
        )

        # 5th and 95th percentiles
        p5 = np.percentile(accuracies, 5)
        print(p5)
        p95 = np.percentile(accuracies, 95)

        ax.set_title(app)
        ax.set_ylim(*y_lim)
        ax.grid(True, linestyle="--", alpha=0.5)
        
        ax.axhline(p5, color="red", linestyle="--", linewidth=2, label="5th percentile")
        ax.axhline(p95, color="green", linestyle="--", linewidth=2, label="95th percentile")

        if idx % y_cols != 0:
            ax.set_ylabel("")

        # IQR annotation (Q3 - Q1)
        q1 = np.percentile(accuracies, 25)
        q3 = np.percentile(accuracies, 75)
        iqr = q3 - q1
        
        ax.text(0.5, 1.15, f"IQR = {iqr:.2f}%", ha="center", fontsize=10, color="blue", transform=ax.transAxes)

        


    # Hide unused subplots
    for j in range(idx + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import os

def collect_5th_percentile_accuracy_by_app(matrix_sizes, base_path="./sensitivity_res/ncf_prediction_accuracy_{}.csv"):
    app_p5_values = {}

    for size in matrix_sizes:
        csv_path = base_path.format(size)
        if not os.path.exists(csv_path):
            print(f"Skipping missing file: {csv_path}")
            continue

        df = pd.read_csv(csv_path)

        # Convert error to accuracy
        df["Prediction Accuracy (%)"] = 100 - df["Prediction Accuracy (%)"]

        # Group by application
        grouped = df.groupby("Application")["Prediction Accuracy (%)"]

        for app, group in grouped:
            p5 = np.percentile(group, 5)  # 5th percentile of accuracy
            if app not in app_p5_values:
                app_p5_values[app] = {}
            app_p5_values[app][f"{size}-App CF"] = p5

    return app_p5_values

def print_app_accuracy_p5_table(app_p5_values, matrix_sizes):
    print("\nTrue 5th Percentile **Prediction Accuracy** (lower tail) by Application:\n")

    # Use your specified matrix_sizes order
    all_sizes = [f"{size}-App CF" for size in matrix_sizes]

    # Print header
    header = "Application".ljust(20) + "".join(size.ljust(15) for size in all_sizes)
    print(header)
    print("-" * len(header))

    # Print rows
    for app in sorted(app_p5_values.keys()):
        row = app.ljust(20)
        for size in all_sizes:
            val = app_p5_values[app].get(size, None)
            row += (f"{val:.2f}%" if val is not None else "-").ljust(15)
        print(row)

def collect_average_accuracy_by_app(matrix_sizes, base_path="./sensitivity_res/ncf_prediction_accuracy_{}.csv"):
    app_avg_values = {}

    for size in matrix_sizes:
        csv_path = base_path.format(size)
        if not os.path.exists(csv_path):
            print(f"Skipping missing file: {csv_path}")
            continue

        df = pd.read_csv(csv_path)

        # Convert error to accuracy
        df["Prediction Accuracy (%)"] = 100 - df["Prediction Accuracy (%)"]

        # Group by application
        grouped = df.groupby("Application")["Prediction Accuracy (%)"]

        for app, group in grouped:
            avg = np.mean(group)
            if app not in app_avg_values:
                app_avg_values[app] = {}
            app_avg_values[app][f"{size}-App CF"] = avg

    return app_avg_values



def print_app_avg_accuracy_table(app_avg_values, matrix_sizes):
    print("\nAverage **Prediction Accuracy** by Application and CF Matrix Size:\n")

    all_sizes = [f"{size}-App CF" for size in matrix_sizes]

    # Print header
    header = "Application".ljust(20) + "".join(size.ljust(15) for size in all_sizes)
    print(header)
    print("-" * len(header))

    # Print each app's row
    for app in sorted(app_avg_values.keys()):
        row = app.ljust(20)
        for size in all_sizes:
            val = app_avg_values[app].get(size, None)
            row += (f"{val:.2f}%" if val is not None else "-").ljust(15)
        print(row)

        


matrix_sizes = [1, 8, 16, 32]
app_p5_values = collect_5th_percentile_accuracy_by_app(matrix_sizes)
print_app_accuracy_p5_table(app_p5_values, matrix_sizes)

app_avg_values = collect_average_accuracy_by_app(matrix_sizes)
print_app_avg_accuracy_table(app_avg_values, matrix_sizes)
```

```python
df = pd.DataFrame(data, columns=['Application', 'CF Method', 'Accuracy', 'CI_Low', 'CI_High'])

# Create pivot for means and CI bounds
pivot_mean = df.pivot(index='Application', columns='CF Method', values='Accuracy')
pivot_low = df.pivot(index='Application', columns='CF Method', values='CI_Low')
pivot_high = df.pivot(index='Application', columns='CF Method', values='CI_High')

# Compute normalized improvement and error bounds over NCF
improvement_records = []
for method in pivot_mean.columns:
    if method == 'NCF':
        continue
    for app in pivot_mean.index:
        ncf_val = pivot_mean.loc[app, 'NCF']
        mean_val = pivot_mean.loc[app, method]
        low_val = pivot_low.loc[app, method]
        high_val = pivot_high.loc[app, method]

        # Improvement and bounds
        improvement = (mean_val - ncf_val) / ncf_val * 100
        low_err = (mean_val - low_val) / ncf_val * 100
        high_err = (high_val - mean_val) / ncf_val * 100

        improvement_records.append([app, method, improvement, low_err, high_err])

# Build DataFrame
improvement_df = pd.DataFrame(improvement_records,
                               columns=['Application', 'CF Method', 'Improvement (%)', 'Low Err', 'High Err'])

# Reorder the applications
app_order = [
    "NAMD", "bert_large", "lammps", "miniGAN", "Resnet50", "sw4lite",
    "UNet", "gromacs", "XSBench", "Laghos"
]


app_order = [
   "miniGAN", "XSBench", "Resnet50",
     "Laghos"
]

colors = ['#c99b38', '#eddca5', '#00b0be', '#8fd7d7']

# Get list of methods
methods = df['CF Method'].unique()
bar_width = 0.1  # slightly wider bars
group_spacing = 0.5  # reduce gap between groups
x = np.arange(len(app_order)) * group_spacing  # shrink horizontal space

plt.figure(figsize=(5.5, 3))

for i, method in enumerate(methods):
    subset = df[df['CF Method'] == method].set_index('Application').loc[app_order]
    accuracies = subset['Accuracy'].values
    yerr = [
        accuracies - subset['CI_Low'].values,
        subset['CI_High'].values - accuracies
    ]

    plt.bar(x + i * bar_width, accuracies, width=bar_width,
            label=method, capsize=3, ecolor='black', lw=1, color=colors[i])

# Adjust x-ticks to match new spacing
plt.xticks(x + bar_width * (len(methods) - 1) / 2 + 0.2, app_order, fontsize=13, rotation=0, ha='right')

plt.yticks(fontsize=18)
plt.ylabel('Prediction Acc. (%)', fontsize=16)
plt.ylim(75, 100)
plt.grid(axis="y")
plt.axhline(100, color='gray', linestyle='--', lw=0.5)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.26), ncol=len(methods), fontsize=11)

plt.tight_layout()
plt.savefig("./fig/sensitivity_algorithm_4.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from itertools import product

applications = [
    "lammps", "Resnet50", "sw4lite", "Laghos", "NAMD",
    "miniGAN", "gromacs", "bert_large", "UNet", "XSBench"
]

# Format: (MLP_acc, MLP_low, MLP_high), (NCF_acc, NCF_low, NCF_high)
combined_data = {
    't=5s': [
        ((97.10, 95.33, 98.86), (97.18, 96.68, 97.68)),
        ((96.79, 95.02, 98.57), (97.48, 96.99, 97.97)),
        ((93.66, 87.58, 99.74), (93.46, 91.78, 95.15)),
        ((96.40, 95.05, 97.76), (96.68, 96.17, 97.19)),
        ((90.54, 82.70, 98.39), (86.45, 83.42, 89.49)),
        ((97.79, 96.46, 99.13), (97.43, 96.89, 97.98)),
        ((97.34, 96.23, 98.44), (97.45, 97.07, 97.84)),
        ((95.72, 92.86, 98.57), (93.07, 91.42, 94.71)),
        ((98.05, 95.89, 100.21), (96.73, 95.94, 97.53)),
        ((94.40, 90.26, 98.54), (92.84, 91.21, 94.46)),
    ],
    't=10s': [
        ((97.48, 96.11, 98.85), (97.48, 96.94, 98.01)),
        ((97.07, 95.63, 98.51), (98.29, 97.85, 98.72)),
        ((94.05, 88.85, 99.25), (93.91, 92.37, 95.45)),
        ((95.98, 94.52, 97.44), (95.72, 95.10, 96.35)),
        ((92.46, 86.63, 98.28), (89.09, 86.81, 91.36)),
        ((97.99, 96.77, 99.21), (96.12, 95.32, 96.92)),
        ((96.52, 94.53, 98.51), (96.57, 95.79, 97.35)),
        ((96.60, 94.13, 99.07), (94.74, 93.41, 96.08)),
        ((97.99, 97.20, 98.77), (97.73, 97.37, 98.09)),
        ((97.04, 95.24, 98.83), (96.65, 95.84, 97.45)),
    ],
    't=15s': [
        ((98.03, 97.35, 98.72), (97.73, 97.38, 98.08)),
        ((96.68, 94.95, 98.41), (96.47, 95.75, 97.19)),
        ((94.19, 88.98, 99.40), (94.20, 92.71, 95.68)),
        ((95.21, 93.40, 97.02), (94.13, 93.23, 95.03)),
        ((92.78, 87.07, 98.48), (89.05, 86.65, 91.45)),
        ((97.77, 96.60, 98.94), (97.55, 97.00, 98.09)),
        ((97.04, 95.94, 98.13), (97.04, 96.52, 97.56)),
        ((96.53, 93.67, 99.39), (93.76, 92.57, 94.95)),
        ((98.28, 97.49, 99.08), (98.47, 98.12, 98.81)),
        ((95.92, 93.30, 98.53), (94.59, 93.41, 95.77)),
    ],
    't=20s': [
        ((98.16, 97.55, 98.77), (98.23, 97.89, 98.57)),
        ((96.18, 93.42, 98.94), (95.40, 94.25, 96.55)),
        ((94.25, 89.03, 99.46), (94.11, 92.62, 95.61)),
        ((95.05, 92.90, 97.21), (93.89, 92.91, 94.88)),
        ((92.78, 87.30, 98.26), (89.00, 86.75, 91.25)),
        ((98.14, 97.08, 99.21), (96.20, 95.43, 96.96)),
        ((97.65, 96.98, 98.31), (97.64, 97.11, 98.17)),
        ((96.65, 93.87, 99.42), (95.16, 94.08, 96.24)),
        ((98.41, 97.62, 99.21), (98.55, 98.23, 98.88)),
        ((96.29, 94.01, 98.57), (94.40, 93.14, 95.66)),
    ]
}

# Combine weighted MLP + NCF scores
records = []
for t, app in product(combined_data.keys(), applications):
    idx = applications.index(app)
    (mlp_acc, mlp_low, mlp_high), (ncf_acc, ncf_low, ncf_high) = combined_data[t][idx]
    acc = 0.1 * mlp_acc + 0.9 * ncf_acc
    low = 0.1 * mlp_low + 0.9 * ncf_low
    high = 0.1 * mlp_high + 0.9 * ncf_high
    records.append([app, t, acc, low, high])

df = pd.DataFrame(records, columns=['Application', 'Profiling Time', 'Accuracy', 'CI_Low', 'CI_High'])
df['Error'] = df['CI_High'] - df['Accuracy']

# Compute relative improvement from t=5s baseline
baseline_acc = df[df['Profiling Time'] == 't=5s'].set_index('Application')['Accuracy']
df['Improvement (%)'] = df.apply(
    lambda row: ((row['Accuracy'] - baseline_acc[row['Application']]) / baseline_acc[row['Application']] * 100)
    if row['Profiling Time'] != 't=5s' else np.nan,
    axis=1
)

# Plotting
apps = df['Application'].unique()
times = ['t=10s', 't=15s', 't=20s']  

bar_width = 0.2
x = np.arange(len(apps))

plt.figure(figsize=(8, 4.5))

for i, t in enumerate(times):
    sub = df[df['Profiling Time'] == t].set_index('Application').loc[apps]
    heights = sub['Improvement (%)'].values
    errors = sub['Error'].values
    plt.bar(x + i * bar_width, heights, width=bar_width, label=t)
    plt.errorbar(x + i * bar_width, heights, yerr=errors, fmt='none', ecolor='black', capsize=3, lw=1)

plt.axhline(0, color='black', linestyle='--', lw=1)
plt.xticks(x + bar_width * (len(times) - 1) / 2, apps, rotation=90, fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Acc. Improvement (%)", fontsize=14)
plt.ylim(-6, 10)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.3), ncol=3, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
ncf_versions = ["2", "8", "16", "20"]
mean_acc = [90.49, 93.89, 94.95, 95.68]

# Interpret these as 98% CI half widths
ci98_half = [4.95/2, 2.90/2, 2.95/2, 2.55/2]

# Symmetric error bars for 98% CI
yerr = np.vstack([ci98_half, ci98_half])

colors = ['#c99b38', '#eddca5', '#00b0be', '#8fd7d7']

# Bar plot with 98% CI
plt.figure(figsize=(4, 2.8))
plt.bar(
    ncf_versions, mean_acc,
    yerr=yerr, capsize=5,
    edgecolor='black', color=colors, width=0.6
)

plt.ylim(75, 100)
plt.ylabel('Avg. Prediction Acc. (%)', fontsize=15)
# plt.xlabel('Initial Dense Matrix Size', fontsize=13)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis="y")
plt.tight_layout()
plt.savefig("./fig/sensitivity_matrix_size.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# CF methods
cf_methods = ['NCF', 'AE', 'SVD', 'KNN']

# Mean accuracy values (your input)
mean_values = [95.13, 93.82, 92.44, 90.92]

# Confidence intervals from grouped data
ci_dict = {
    'NCF': (94.081, 96.922),
    'AE': (92.697, 94.923),
    'SVD': (90.399, 94.384),
    'KNN': (88.489, 93.271),
}

# Build asymmetric error bars
ci_low = [mean_values[i] - ci_dict[cf_methods[i]][0] for i in range(len(cf_methods))]
ci_high = [ci_dict[cf_methods[i]][1] - mean_values[i] for i in range(len(cf_methods))]
yerr = [ci_low, ci_high]

# Colors
colors = ['#c99b38', '#eddca5', '#00b0be', '#8fd7d7']

# Plot
plt.figure(figsize=(2.7, 3))
plt.bar(cf_methods, mean_values, 
        yerr=yerr, capsize=5, 
        edgecolor='black', color=colors, width=0.7)

plt.ylim(65, 100)
plt.ylabel('Avg. Prediction Acc. (%)', fontsize=15)
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=14)
plt.grid(axis="y")
plt.tight_layout()
plt.savefig("./fig/sensitivity_algorithm.png", dpi=300, bbox_inches='tight')
plt.show()
```
